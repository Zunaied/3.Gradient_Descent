# Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x,y):
    m_curr=b_curr=0
    n=len(x)
    learning_rate=0.05
    iterations=10
    plt.scatter(x,y,color='red',marker='*',linewidths='5')
    
    for i in range(iterations):
        yp=m_curr*x+b_curr
        %matplotlib inline
        plt.plot(x,yp,color='green')
        cost=(1/n)*sum((y-yp))**2
        md=-(2/n)*sum(x*(y-yp))
        bd=-(2/n)*sum((y-yp))
        m_curr=m_curr-learning_rate*md
        b_curr=b_curr-learning_rate*bd
        
        
        print("m=",m_curr,"\t","b=",b_curr,"\t","cost=",cost)
    
   
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
gradient_descent(x,y)
```
